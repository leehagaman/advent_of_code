```python
import numpy as np

from tqdm.notebook import tqdm

import matplotlib.pyplot as plt

from queue import PriorityQueue
```

```python
with open("input_day_24.txt") as file:
    
    text = file.read()
```

```python
test = False
if test:

    text = """
#.######
#>>.<^<#
#.<..<<#
#>v.><>#
#<^v^^>#
######.#
"""
    
text = text.strip()
```

```python
no_wall_text = ""
for l in text.split("\n")[1:-1]:
    no_wall_text += "\n"
    no_wall_text += l[1:-1]
    
no_wall_text = no_wall_text.strip()

y_h = len(no_wall_text.split("\n"))
x_h = len(no_wall_text.split("\n")[0])

print(x_h, y_h)
#print(no_wall_text)
```

```python
b_positions = set()
for y, l in enumerate(no_wall_text.split("\n")):
    for x, c in enumerate(l):
        if c != ".":
            b_positions.add(((x, y), c))
```

```python
empty_pos_each_min = []

for minute in tqdm(range(1000)):
    
    b_positions_set = set()
    for b_pos in b_positions:
        b_positions_set.add(b_pos[0])
        
    #print("not empty:", b_positions_set)
    
    empty_set = set()
    for y in range(y_h):
        for x in range(x_h):
            if not ((x, y) in b_positions_set):
                empty_set.add((x, y))
    #print("empty:", empty_set)
    empty_pos_each_min.append(empty_set)
    
    new_b_positions = set()
    for b_pos in b_positions.copy():
        pos, d = b_pos
        if d == ">":
            new_pos = ((pos[0] + 1) % x_h, pos[1])
        elif d == "<":
            new_pos = ((pos[0] - 1) % x_h, pos[1])
        elif d == "^":
            new_pos = (pos[0], (pos[1] - 1) % y_h)            
        elif d == "v":
            new_pos = (pos[0], (pos[1] + 1) % y_h) 
        
        new_b_positions.add((new_pos, d))
        
    b_positions = new_b_positions
        
    
```

```python
do_part_1 = False

if do_part_1:
    q = PriorityQueue()

    def estimate_dist(state):
        x, y = state[0]
        steps = state[1]
        return steps + x - x_h + y - y_h

    # position, minutes passed
    # we assume that we move to the top left corner on the first step, checked inputs allow this
    current_state = ((0, 0), 0)

    q.put((estimate_dist(current_state), current_state))

    prev_considered_state_hashes = set()

    min_num_steps = 1e9

    num_tries = -1

    while not q.empty():

        #print(q.qsize())

        num_tries += 1

        if test:
            if num_tries % 1000 == 0:
                print(num_tries, min_num_steps + 1)
        else:
            if num_tries % 100000 == 0:
                print(num_tries, min_num_steps + 1)

        estimated_dist, current_state = q.get()

        current_hash = hash(current_state)
        if current_hash in prev_considered_state_hashes:
            continue
        else:
            prev_considered_state_hashes.add(current_hash)

        pos, num_minutes = current_state

        if pos == (x_h - 1, y_h - 1):
            #print("found solution")
            min_num_steps = min(min_num_steps, num_minutes)
            continue

        for d in ((0, 0), (1, 0), (0, 1), (-1, 0), (0, -1)):

            #print("trying to add steps")

            next_pos = (pos[0] + d[0], pos[1] + d[1])
            if next_pos in empty_pos_each_min[num_minutes + 1]:
                #print("adding step")
                next_state = (next_pos, num_minutes + 1)
                q.put((estimate_dist(next_state), next_state))

    print("exhausted all possibilities")
```

```python
q = PriorityQueue()

def estimate_dist(state):
    #print(state)
    x, y = state[0]
    mins = state[1]
    step = state[2]
    
    if step == 0:
        return mins + (x - x_h + y - y_h) + 2 * (x_h + y_h)
    elif step == 1:
        return mins + 2 * (x_h + y_h)
    elif step == 2:
        return mins + (x + y) + (x_h + y_h)
    elif step == 3:
        return mins + (x_h + y_h)
    elif step == 4:
        return mins + (x - x_h + y - y_h)
    else:
        print("invalid step!")

# position, minutes passed, step
# step 0 is going to goal, step 1 is in the goal, step 2 is going to start, step 3 is in the start, step 4 is going to goal
# we assume that we move to the top left corner on the first step, checked inputs allow this
current_state = ((0, 0), 1, 0)
    
q.put((estimate_dist(current_state), current_state))

prev_considered_state_hashes = set()

min_num_steps = 1e9

num_tries = -1

while not q.empty():
    
    #print(q.qsize())
        
    num_tries += 1
    
    if test:
        if num_tries % 1000 == 0:
            print(num_tries, min_num_steps)
    else:
        if num_tries % 100000 == 0:
            print(num_tries, min_num_steps)
            
    estimated_dist, current_state = q.get()
    
    current_hash = hash(current_state)
    if current_hash in prev_considered_state_hashes:
        continue
    else:
        prev_considered_state_hashes.add(current_hash)
        
    pos, num_minutes, step = current_state
    
    if step == 0 and pos == (x_h - 1, y_h - 1): # going into goal the first time
        next_state = (pos, num_minutes + 1, 1)
        q.put((estimate_dist(next_state), next_state))
        continue
        
    if step == 1: # waiting at goal
        if pos in empty_pos_each_min[num_minutes]:
            next_state = (pos, num_minutes + 1, 2)
            q.put((estimate_dist(next_state), next_state))
        else:
            next_state = (pos, num_minutes + 1, 1)
            q.put((estimate_dist(next_state), next_state))        
        continue
        
    if step == 2 and pos == (0, 0): # going into goal the first time
        next_state = (pos, num_minutes + 1, 3)
        q.put((estimate_dist(next_state), next_state))
        continue
        
    if step == 3: # waiting at goal
        if pos in empty_pos_each_min[num_minutes]:
            next_state = (pos, num_minutes + 1, 4)
            q.put((estimate_dist(next_state), next_state))
        else:
            next_state = (pos, num_minutes + 1, 3)
            q.put((estimate_dist(next_state), next_state))        
        continue
        
    if step == 4 and pos == (x_h - 1, y_h - 1):
        if num_minutes + 1 < min_num_steps:
            min_num_steps = num_minutes
            print("new best: ", min_num_steps)
            continue
        
    for d in ((0, 0), (1, 0), (0, 1), (-1, 0), (0, -1)):
        
        #print("trying to add steps")
        
        next_pos = (pos[0] + d[0], pos[1] + d[1])
        if next_pos in empty_pos_each_min[num_minutes]:
            #print("adding step")
            next_state = (next_pos, num_minutes + 1, step)
            q.put((estimate_dist(next_state), next_state))
    
print("exhausted all possibilities")
```
