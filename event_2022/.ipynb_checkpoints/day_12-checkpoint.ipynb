```python
import numpy as np

from tqdm.notebook import tqdm
```

```python
#with open("input_day_12_test.txt") as file:
with open("input_day_12.txt") as file:
    
    text = file.read()
```

```python
test = False

if test:
    text = """
    Sabqponm
    abcryxxl
    accszExk
    acctuvwj
    abdefghi
    """
```

```python
arr = [[x for x in l.strip()] for l in text.strip().split("\n")]

#print(arr)

for r in range(len(arr)):
    for c in range(len(arr[0])):
        if arr[r][c] == "S":
            start_r = r
            start_c = c
            
        if arr[r][c] == "E":
            end_r = r
            end_c = c
```

```python
slow = False

if slow:

    cum_arr = [[10000000 for c in range(len(arr[0]))] for r in range(len(arr))]

    cum_arr[start_r][start_c] = 0 # starting position

    for t in tqdm(range(10000)): # updating to make it better
        for r in range(len(arr)):
            for c in range(len(arr[0])):
                # iterate over directions to look
                new_vals = [cum_arr[r][c]] # keeping track of the possible min steps for this location
                for dr, dc in [[-1, 0], [0, -1], [1, 0], [0, 1]]:
                    if r + dr < 0 or r + dr > len(arr) - 1: # stepping off the edge
                        continue
                    if c + dc < 0 or c + dc > len(arr[0]) - 1: # stepping off the edge
                        continue
                    if ord(arr[r + dr][c + dc]) - ord(arr[r][c]) > 1: # stepping too high
                        continue
                    #print(r, r + dr, c, c + dc)
                    #print(new_vals, cum_arr[r + dr][c + dc] + 1)
                    new_vals.append(cum_arr[r + dr][c + dc] + 1) # one step more than other locations

                #print(new_vals)
                cum_arr[r][c] = np.min(new_vals)
                if cum_arr[end_r][end_c] < 10000000:
                    print(cum_arr[end_r][end_c])

        #print(cum_arr)
        
    print(cum_arr[end_r][end_c])
    
    for r in cum_arr:
        print("\n")
        for c in r:
            if c < 1e9:
                print(str(int(c))+",", end = "")
            else:
                print("-", end="")
                
```

```python
from queue import PriorityQueue

q = PriorityQueue()


q.put((abs(start_r - end_r) + abs(start_c - end_c), [start_r, start_c])) # putting the first point with estimated cost


cum_steps = [[10000000 for c in range(len(arr[0]))] for r in range(len(arr))] # not evaluated, assume maximum

cum_steps[start_r][start_c] = 0

#print(q.get())

while not q.empty():
    
    #print(q.queue)
    
    e = q.get()
    #print(e)
    #print("new size", q.qsize())

    #print(e)
    r = e[1][0]
    c = e[1][1]
    
    #print("frontier point", r, c)
    #print(len(big_arr) - 1, len(big_arr[0]) - 1)
    
    if r == end_r and c == end_c:
        print("got it, breaking")
        break
    
    for dr, dc in [[-1, 0], [0, -1], [1, 0], [0, 1]]:
        
        #print(dr, dc)
        
        if r + dr < 0 or r + dr > len(arr) - 1:
            continue
        if c + dc < 0 or c + dc > len(arr[0]) - 1:
            continue
            
        #print("here", arr[r][c], "to", arr[r + dr][c + dc])
        if arr[r][c] == "S":
            pass
            #print("here1")

        elif arr[r + dr][c + dc] == "E" and arr[r][c] == "z":
            pass
            #print("here2")

        elif arr[r + dr][c + dc] == "E":
            #print("here3")
            continue

        elif ord(arr[r + dr][c + dc]) - ord(arr[r][c]) > 1: # appears to be too high
            #print("here4")
            continue

        #print("here5")
        
        #print("new and old steps", cum_steps[r][c] + 1, cum_steps[r + dr][c + dc])
        
        if cum_steps[r][c] + 1 < cum_steps[r + dr][c + dc]: # improved, add to frontier
            
            #print("improved", cum_steps[r][c] + 1, "<",  cum_steps[r + dr][c + dc])
            
            cum_steps[r + dr][c + dc] = cum_steps[r][c] + 1

            q.put((cum_steps[r + dr][c + dc] + len(arr) - r + len(arr[0]) - c, [r + dr, c + dc]))
            
print("done!")
print(cum_steps[end_r][end_c])
```
