```python
def print_map(map_arr, visited_set):
    for y in range(len(map_arr)):
        s = ""
        for x in range(len(map_arr[0])):
            if (x, y) in visited_set:
                s += "*"
            else:
                s += map_arr[y][x]
                
        print(s)
    
    
```
