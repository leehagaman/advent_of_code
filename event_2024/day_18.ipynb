```python
import numpy as np
import copy

import heapq

from collections import defaultdict

from tqdm.notebook import tqdm

import matplotlib.pyplot as plt
```

```python
with open('input_day_18.txt', 'r') as file:
    input_string = file.read()


test_input_string = """
5,4
4,2
4,5
3,0
2,1
6,3
2,4
1,5
0,6
3,3
2,6
5,1
1,2
5,5
2,5
6,5
1,4
0,4
6,4
1,1
6,1
1,0
0,5
1,6
2,0
""".strip()

test = False

if test:
    num_bytes = 12
    R, C = 7, 7
    input_string = test_input_string
else:
    num_bytes = 1024
    R, C = 71, 71
```

```python
positions = []
for b in range(num_bytes):
    pair = input_string.split("\n")[b]
    positions.append([int(c) for c in pair.split(",")])

grid = [["." for c in range(C)] for r in range(R)]

for pos in positions:
    c, r = pos
    grid[r][c] = "#"

def estimate_cost(state):
    pos, cost = state
    pos_c, pos_r = pos
    return cost + (R-1) - pos_r + (C-1) - pos_c

best_cost = 1000000000

q = []
heapq.heapify(q)
start_state = ((0, 0), 0)
heapq.heappush(q, (estimate_cost(start_state), start_state))

positions_checked = set()

num_states_considered = 0

while len(q) > 0:
    est_cost, state = heapq.heappop(q)
    pos, cost = state
    pos_c, pos_r = pos

    #if num_states_considered > 20:
    #    break

    
    positions_checked.add(pos)

    if num_states_considered % 100_000 == 0:
        print(num_states_considered, est_cost, state, len(q), len(positions_checked))
    num_states_considered += 1

    if pos_r == R-1 and pos_c == C-1:
        best_cost = min(best_cost, cost)
        print("success!!", best_cost)

    for new_dir in ((0, 1), (0, -1), (1, 0), (-1, 0)):
        new_pos_c = pos_c + new_dir[0]
        new_pos_r = pos_r + new_dir[1]

        if not (0 <= new_pos_c < C and 0 <= new_pos_r < R):
            continue

        if grid[new_pos_r][new_pos_c] == ".":
            new_state = ((new_pos_c, new_pos_r), cost + 1)
            if (new_pos_c, new_pos_r) not in positions_checked:
                #print("(new_pos_c, new_pos_r) not in positions_checked,", (new_pos_c, new_pos_r), positions_checked)
                if (estimate_cost(new_state), new_state) not in q:
                    heapq.heappush(q, (estimate_cost(new_state), new_state))
```
