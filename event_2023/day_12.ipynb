```python
import numpy as np
import math
from tqdm.notebook import tqdm
from functools import cache
```

```python
with open("input_day_12.txt") as f:
    text = f.read()


test = False
if test:
    text = """
???.### 1,1,3
.??..??...?##. 1,1,3
?#?#?#?#?#?#?#? 1,3,1,6
????.#...#... 4,1,1
????.######..#####. 1,6,5
?###???????? 3,2,1
"""
    text = text.strip()
```

```python
@cache
def get_num_legal(labels, counts):#, num_spaces=0, max_depth=-1, printout=False):
    #if printout: print(" "*num_spaces + "in function,", labels, counts)

    #if max_depth == 0:
    #    return -99999999

    havent_reached_yet = True
    reached_question = False
    all_questions = True
    min_first_num_continuous = 0
    max_first_num_continuous = 0
    end_index_first_num_continuous = 0
    for i, c in enumerate(labels):
        #print("curr: ", c, min_first_num_continuous, max_first_num_continuous)
        if c == "." and havent_reached_yet: # before the first continuous section, keep looking
            continue
        if c == "#": # expand the continuous section, keep looking
            all_questions = False
            havent_reached_yet = False
            if not reached_question:
                min_first_num_continuous += 1
            max_first_num_continuous += 1
        if c == "?":
            reached_question = True
            havent_reached_yet = False
            max_first_num_continuous += 1
        if c == ".": # done with continuous section, stop
            end_index_first_num_continuous = i
            break
        if i == len(labels) - 1: # reached end of line, stop
            end_index_first_num_continuous = i + 1
            break

    #if printout: print(" "*num_spaces, min_first_num_continuous, max_first_num_continuous)
    
    if min_first_num_continuous != max_first_num_continuous: # uncertain number, probably have to recurse
        if not all_questions and (min_first_num_continuous > counts[0] or max_first_num_continuous < counts[0]): # impossible, no need to recurse
            #if printout: print(" "*num_spaces, "impossible, min, max, needed:", min_first_num_continuous, max_first_num_continuous, counts[0])
            return 0
    elif min_first_num_continuous != counts[0]: # already invalid, none possible
        #if printout: print(" "*num_spaces + "first continuous invalid")
        return 0
    elif len(counts) == 1: # valid and only one length, done, one possible
        if "#" in labels[end_index_first_num_continuous:]: # there's another thing not accounted for
            #if printout: print(" "*num_spaces + "another one not accounted for, section examined: ", labels[end_index_first_num_continuous:])
            return 0
        #if printout: print(" "*num_spaces + "  good")
        return 1
    else:
        return get_num_legal(labels[end_index_first_num_continuous:], counts[1:])#, num_spaces=num_spaces + 2, max_depth=max_depth-1, printout=printout) # first segment valid, can look again to find the next continuous segment
        

    test1 = labels.replace("?", '.', 1)
    test2 = labels.replace("?", '#', 1)
    return get_num_legal(test1, counts) + get_num_legal(test2, counts)#, num_spaces=num_spaces + 2, max_depth=max_depth-1, printout=printout)
    
```

```python
nums_legal = []
lines = text.split("\n")
for line in lines:
    labels, counts = line.split(" ")
    counts = tuple([int(x) for x in counts.split(",")])

    print(type(labels), type(counts), type(nums_legal), type(get_num_legal))

    nums_legal.append(get_num_legal(labels, counts))

print(nums_legal)
print(sum(nums_legal))
    

```
