```python
with open("input_day_22.txt") as f:
    text = f.read()


test = False
if test:
    text = r"""
1,0,1~1,2,1
0,0,2~2,0,2
0,2,3~2,2,3
0,0,4~0,2,4
2,0,5~2,2,5
0,1,6~2,1,6
1,1,8~1,1,9
"""

text = text.strip()
```

```python
floating_bricks = []
for line in text.split("\n"):
    coord1, coord2 = line.split("~")
    floating_bricks.append(
        (
            tuple([int(_) for _ in coord1.split(",")]),
            tuple([int(_) for _ in coord2.split(",")])
        )
    )

floating_bricks.sort(key=lambda b: min(b[0][2], b[1][2]))
```

```python
bricks = copy.deepcopy(floating_bricks)
for brick in copy.deepcopy(bricks):

    occupied_voxels = set()
    for brick2 in bricks:

        if brick2 == brick:
            continue

        diff_x = abs(brick2[0][0] - brick2[1][0])
        diff_y = abs(brick2[0][1] - brick2[1][1])
        diff_z = abs(brick2[0][2] - brick2[1][2])

        min_x = min(brick2[0][0], brick2[1][0])
        min_y = min(brick2[0][1], brick2[1][1])
        min_z = min(brick2[0][2], brick2[1][2])

        for x in range(min_x, min_x + diff_x + 1):
            for y in range(min_y, min_y + diff_y + 1):
                for z in range(min_z, min_z + diff_z + 1):
                    occupied_voxels.add((x,y,z))

    #print(occupied_voxels)


    #print("brick:", brick)
    coord_1 = list(brick[0])
    coord_2 = list(brick[1])
    done_falling = False
    while not done_falling:
        coord_1[2] -= 1
        coord_2[2] -= 1

        diff_x = abs(coord_1[0] - coord_2[0])
        diff_y = abs(coord_1[1] - coord_2[1])
        diff_z = abs(coord_1[2] - coord_2[2])
        
        min_x = min(coord_1[0], coord_2[0])
        min_y = min(coord_1[1], coord_2[1])
        min_z = min(coord_1[2], coord_2[2])

        for x in range(min_x, min_x + diff_x + 1):
            for y in range(min_y, min_y + diff_y + 1):
                for z in range(min_z, min_z + diff_z + 1):
                    if z <= 0 or ((x,y,z) in occupied_voxels):
                        #print(x,y,z-1, "problem", coord_1, coord_2)
                        done_falling = True

    coord_1[2] += 1
    coord_2[2] += 1
    bricks.remove(brick)
    bricks.append((tuple(coord_1), tuple(coord_2)))
    #print("fell to:", (tuple(coord_1), tuple(coord_2)))
```

```python
def intersecting(brick1, brick2):
    
    diff_x_1 = abs(brick1[0][0] - brick1[1][0])
    diff_y_1 = abs(brick1[0][1] - brick1[1][1])
    diff_z_1 = abs(brick1[0][2] - brick1[1][2])

    diff_x_2 = abs(brick2[0][0] - brick2[1][0])
    diff_y_2 = abs(brick2[0][1] - brick2[1][1])
    diff_z_2 = abs(brick2[0][2] - brick2[1][2])

    min_x_1 = min(brick1[0][0], brick1[1][0])
    min_y_1 = min(brick1[0][1], brick1[1][1])
    min_z_1 = min(brick1[0][2], brick1[1][2])

    min_x_2 = min(brick2[0][0], brick2[1][0])
    min_y_2 = min(brick2[0][1], brick2[1][1])
    min_z_2 = min(brick2[0][2], brick2[1][2])

    #print("brick 1 diffs:", diff_x_1, diff_y_1, diff_z_1)
    #print("brick 2 diffs:", diff_x_2, diff_y_2, diff_z_2)

    #print("min brick 1s:", min_x_1, min_y_1, min_z_1)
    #print("min brick 2s:", min_x_2, min_y_2, min_z_2)

    occupied_set = set()
    for x in range(min_x_1, min_x_1 + diff_x_1 + 1):
        for y in range(min_y_1, min_y_1 + diff_y_1 + 1):
            for z in range(min_z_1, min_z_1 + diff_z_1 + 1):
                #print("here, x, y, z")
                occupied_set.add((x,y,z))

    #print("occupied set", occupied_set)

    for x in range(min_x_2, min_x_2 + diff_x_2 + 1):
        for y in range(min_y_2, min_y_2 + diff_y_2 + 1):
            for z in range(min_z_2, min_z_2 + diff_z_2 + 1):
                if (x,y,z) in occupied_set:
                    #print("intersecting block,", x,y,z)
                    return True
    
    return False
```

```python
# brick : below_bricks, above_bricks
brick_dic = {}
for brick1 in tqdm(bricks):
    below_bricks = []
    above_bricks = []
    below_brick1 = ((brick1[0][0], brick1[0][1], brick1[0][2]-1), (brick1[1][0], brick1[1][1], brick1[1][2]-1))
    above_brick1 = ((brick1[0][0], brick1[0][1], brick1[0][2]+1), (brick1[1][0], brick1[1][1], brick1[1][2]+1))

    #print(brick1)
    #print(below_brick1)
    #print(above_brick1)
    for brick2 in bricks:
        if brick2 == brick1:
            continue
        if intersecting(below_brick1, brick2):
            below_bricks.append(brick2)
        if intersecting(above_brick1, brick2):
            above_bricks.append(brick2)

    brick_dic[brick1] = [below_bricks, above_bricks]
        
```

```python
def num_falling(removed_bricks, p=""):

    new_removed_bricks = copy.deepcopy(removed_bricks)

    #print(p + "removing", removed_bricks)

    ret = 0

    all_above_bricks = []
    for removed_brick in removed_bricks:
        all_above_bricks += brick_dic[removed_brick][1]

    for above_brick in all_above_bricks:
        if above_brick in new_removed_bricks:
            continue
        supporting_bricks = brick_dic[above_brick][0]
        one_support_staying = False
        for supporting_brick in supporting_bricks:
            if not (supporting_brick in removed_bricks):
                one_support_staying = True
        if not one_support_staying:
            new_removed_bricks.append(above_brick)

    
    if new_removed_bricks == removed_bricks:
        return len(removed_bricks)
    
    return num_falling(new_removed_bricks)

tot = 0
for  brick in tqdm(bricks):

    num_that_would_fall = num_falling([brick]) - 1
    tot += num_that_would_fall
    #print(num_that_would_fall)

print("tot:", tot)
```
