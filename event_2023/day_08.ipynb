```python
import math

def lcm(a, b):
    return abs(a*b) // math.gcd(a, b)
```

```python
with open("input_day_08.txt") as f:
    text = f.read()


test = False
if test:
    text = """RL

AAA = (BBB, CCC)
BBB = (DDD, EEE)
CCC = (ZZZ, GGG)
DDD = (DDD, DDD)
EEE = (EEE, EEE)
GGG = (GGG, GGG)
ZZZ = (ZZZ, ZZZ)"""

test2 = False
if test2:
    text = """LLR

AAA = (BBB, BBB)
BBB = (AAA, ZZZ)
ZZZ = (ZZZ, ZZZ)"""


test3 = False
if test3:
    text = """LR

11A = (11B, XXX)
11B = (XXX, 11Z)
11Z = (11B, XXX)
22A = (22B, XXX)
22B = (22C, 22C)
22C = (22Z, 22Z)
22Z = (22B, 22B)
XXX = (XXX, XXX)"""
```

```python
dirs = text.split("\n\n")[0]
lines = text.split("\n\n")[1].split("\n")
```

```python
dic = {}
for line in lines:
    root = line.split(" = ")[0]
    left = line.split(" = ")[1].split(", ")[0][1:]
    right = line.split(" = ")[1].split(", ")[1][:-1]
    dic[root] = (left, right)
```

```python
i = 0
curr = "AAA"
while True:

    d = dirs[i%len(dirs)]

    if d == "L":
        curr = dic[curr][0]
    else:
        curr = dic[curr][1]

    if curr == "ZZZ":
        print(i+1)
        break


    i += 1

```

```python
cycle_dic = {}
for line in lines:
    if line[2] == "A":
        first = line[:3]
        curr = first
        print("new first:", first)
        i = 0

        while True:

            d = dirs[i%len(dirs)]

            print(d, curr)

            if d == "L":
                curr = dic[curr][0]
            else:
                curr = dic[curr][1]

            if curr[2] == "Z":
                cycle_dic[first] = i+1
                print("done")
                break


            i += 1

mul = 1
for k, v in cycle_dic.items():
    mul = lcm(mul, v)
print(mul)
```
