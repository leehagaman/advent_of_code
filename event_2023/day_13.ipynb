```python
with open("input_day_13.txt") as f:
    text = f.read()


test = False
if test:
    text = """
#.##..##.
..#.##.#.
##......#
##......#
..#.##.#.
..##..##.
#.#.##.#.

#...##..#
#....#..#
..##..###
#####.##.
#####.##.
..##..###
#....#..#
"""
    text = text.strip()

test2 = False
if test2:
    text = """
.#.#....#.#.###
...#....#...#..
....####....#..
.#.######.#..##
###..##..###...
.#...##...#.###
..##....##.....
.#...##...#..##
..##....##..#..
.##########....
##.#.##.#.##...
....####.....##
##...##...##.##
##.#.##.#.##.##
...#....#...###
##...##...#####
....#..........
"""
    text = text.strip()

tests = [t.split("\n") for t in text.split("\n\n")]
```

```python
horizontals = []
verticals = []

num = 0

detailed_print = False

for t in tests:
    if detailed_print: print("new test")
    horizontals.append([])
    verticals.append([])
    R = len(t)
    C = len(t[0])

    for r in range(1, R):
        if detailed_print: print("  checking new horizontal", r)
        works = True
        for rr in range(min(r, R - r)):
            if detailed_print: print("    checking rows: ")
            if detailed_print: print("      ", t[r - rr - 1])
            if detailed_print: print("      ", t[r + rr])
            if t[r - rr - 1] != t[r + rr]:
                works = False
                if detailed_print: print("        not a reflection!")
                break
        if works:
            horizontals[-1].append(r)
            num += r * 100

    t_transpose = []
    for c in range(C):
        t_transpose.append("")
        for r in range(R):
            t_transpose[-1] += t[r][c]

    for c in range(1, C):
        if detailed_print: print("\n  checking new vertical", c)
        works = True
        for cc in range(min(c, C - c)):
            if detailed_print: print("    checking cols: ")
            if detailed_print: print("      ", t_transpose[c - cc - 1])
            if detailed_print: print("      ", t_transpose[c + cc])
            if t_transpose[c - cc - 1] != t_transpose[c + cc]:
                works = False
                if detailed_print: print("        not a reflection!")
                break
        if works:
            verticals[-1].append(c)
            num += c

print(num)
```
