```python
lines = []
lines.append("."*(len(lines_[0]) + 4))
lines.append("."*(len(lines_[0]) + 4))
for l in lines_:
    #print(len(l))
    lines.append(".." + l + "..")
lines.append("."*(len(lines_[0]) + 4))
lines.append("."*(len(lines_[0]) + 4))
```

```python
tot = 0
curr_num = ""
number_active = False
for r in range(1, len(lines) - 1):
    for c in range(1, len(lines[0]) - 1):
        if lines[r][c].isdigit(): # part of a digit
            curr_num += lines[r][c]
            for dir in [(-1, -1),
                        (-1, 0),
                        (-1, 1),
                        (0, -1),
                        (0, 0),
                        (0, 1),
                        (1, -1),
                        (1, 0),
                        (1, 1)]:
                test_char = lines[r + dir[0]][c + dir[1]]
                if not (test_char == "." or test_char.isdigit()): # current digit connected to a symbol
                    number_active = True
        else: # current digit is finished
            if number_active:
                tot += int(curr_num) # active number, adding to total
            curr_num = "" # resetting for new number
            number_active = False
        
tot
```
