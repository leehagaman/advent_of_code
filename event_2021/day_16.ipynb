```python
import numpy as np
import copy
```

```python
with open("input_day_16.txt") as file:
    
    s = file.read().strip()
    
    #s = "D2FE28" # for testing, literal 2021
    #s = "8A004A801A8002F478" # for testing, type id 0 with 2 subpackets
    
    #s = "04005AC33890" # for testing

    num = int(s, 16)
    
    bin_str = "{0:b}".format(num).zfill(len(s) * 4)


print(bin_str)
```

```python
tot_version_num = 0

def parse_bin_str(s):
    
    #print("\nnew packet")
    #print(s)
    
    #local num_bits
    
    global tot_version_num
    
    packet_version = int(s[:3], 2)
    
    tot_version_num += packet_version
    
    type_id = int(s[3:6], 2)
    
    #print("packet version", packet_version)
    #print("type id", type_id)
    
    if type_id == 4: # literal value
        lit_val_str = ""
        i = 6
        #print(s)
        while s[i] == "1":
            lit_val_str += s[i + 1 : i + 5]
            i = i + 5
        lit_val_str += s[i + 1 : i + 5]
            
        lit_val = int(lit_val_str, 2)
            
        #print("literal value", lit_val)
        
        #print("packet length", i + 5)
        return lit_val, i + 5 # value, length
        
    else: # operator
        
        #print("operator")
        
        length_type_id = s[6]
        
        #print("length_type_id", length_type_id)
        
        contained_vals = []
        
        if length_type_id == "0": # showing num bits
            num_bits = int(s[7 : 7 + 15], 2)
            #print("first num bits", num_bits)
            i = 7 + 15
            
            while num_bits > 0:
                #print("in here")
                val, recur_len = parse_bin_str(s[i:])
                contained_vals.append(val)
                #print("length of recursive step", i)
                #print("new num bits after recursion", num_bits)
                
                i += recur_len
                num_bits -= recur_len
                #print("new num bits after subtraction", num_bits)
            
            #print("packet length", i)
        
        elif length_type_id == "1": # showing num packets
            num_packets = int(s[7 : 7 + 11], 2)
            i = 7 + 11
            #print("num packets", num_packets)
            
            while num_packets > 0:
                val, recur_len = parse_bin_str(s[i:])
                i += recur_len
                contained_vals.append(val)

                num_packets -= 1
                
            #print("packet length", i)
        
        
        if type_id == 0: # sum
            val = 0
            for v in contained_vals:
                val = val + v
        elif type_id == 1: # product
            val = 1
            for v in contained_vals:
                val = val * v
        elif type_id == 2: # minimum
            val = np.min(contained_vals)
        elif type_id == 3: # minimum
            val = np.max(contained_vals)
        elif type_id == 5: # greater
            val = (contained_vals[0] > contained_vals[1])
        elif type_id == 6: # less
            val = (contained_vals[0] < contained_vals[1])
        elif type_id == 7: # equal
            val = (contained_vals[0] == contained_vals[1])
                
        
        
        return val, i
            
    
val = parse_bin_str(bin_str)[0]

print("\n\ntotal version: ", tot_version_num)
print("result: ", val)
```
