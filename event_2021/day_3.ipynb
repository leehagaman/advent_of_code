```python
import numpy as np
```

```python
bin_nums = np.genfromtxt("input_day_4.txt", dtype=None)
bin_nums[:5]
```

```python
gamma = 0
epsilon = 0

nums_0 = []
nums_1 = []

for pos in range(12):
    nums_0.append(0)
    nums_1.append(0)
    for x in bin_nums:
        k = int(str(x).zfill(12)[pos])
        #print(k)
        if k == 0:
            nums_0[-1] += 1
        elif k == 1:
            nums_1[-1] += 1
        else:
            print("bad!")
            
for pos in range(12):
    if nums_1[pos] > nums_0[pos]:
        gamma += 1 * 2**(11-pos)
        
for pos in range(12):
    if nums_1[pos] < nums_0[pos]:
        epsilon += 1 * 2**(11-pos)


print(nums_0)
print(nums_1)
        
print(gamma, epsilon)
        
print(gamma * epsilon)
```

```python
# part 2

oxy_rating = 0
co2_rating = 0

prev_oxy_possibilities = np.copy(bin_nums)
next_oxy_possibilities = []

prev_co2_possibilities = np.copy(bin_nums)
next_co2_possibilities = []

for pos in range(12):
    
    #print("length:", pos, len(prev_oxy_possibilities))

    nums_0 = 0
    nums_1 = 0

    for i, x in enumerate(prev_oxy_possibilities):

        k = int(str(x).zfill(12)[pos])

        if k == 0:
            nums_0 += 1
        elif k == 1:
            nums_1 += 1

    print("pos, prev, next:", pos, len(prev_oxy_possibilities), len(next_oxy_possibilities))
    
    for i, x in enumerate(prev_oxy_possibilities):

        k = int(str(x).zfill(12)[pos])

        if (nums_1 >= nums_0 and k == 0) or (nums_1 < nums_0 and k == 1):
            continue

        next_oxy_possibilities.append(x)
        
    #print("second loop done")
            
    if len(next_oxy_possibilities) == 1:
        oxy_rating = next_oxy_possibilities[0]
        print("oxy rating is", oxy_rating)
        break

    prev_oxy_possibilities = np.copy(next_oxy_possibilities)
    next_oxy_possibilities = []
    
    
    
for pos in range(12):
    
    #print("length:", pos, len(prev_co2_possibilities))

    nums_0 = 0
    nums_1 = 0

    for i, x in enumerate(prev_co2_possibilities):

        k = int(str(x).zfill(12)[pos])

        if k == 0:
            nums_0 += 1
        elif k == 1:
            nums_1 += 1

    print("pos, prev, next:", pos, len(prev_co2_possibilities), len(next_co2_possibilities))
    
    for i, x in enumerate(prev_co2_possibilities):

        k = int(str(x).zfill(12)[pos])

        if (nums_1 >= nums_0 and k == 1) or (nums_1 < nums_0 and k == 0):
            continue

        next_co2_possibilities.append(x)
        
    #print("second loop done")
            
    if len(next_co2_possibilities) == 1:
        co2_rating = next_co2_possibilities[0]
        print("co2 rating is", co2_rating)
        break

    prev_co2_possibilities = np.copy(next_co2_possibilities)
    next_co2_possibilities = []
    

oxy_num = 0
for i, k in enumerate([int(str(oxy_rating).zfill(12)[_]) for _ in range(12)]):
    if k == 1:
        oxy_num += k * 2**(11 - i)

co2_num = 0
for i, k in enumerate([int(str(co2_rating).zfill(12)[_]) for _ in range(12)]):
    if k == 1:
        co2_num += k * 2**(11 - i)

print(oxy_rating, co2_rating)    

print(oxy_num * co2_num)    
```
