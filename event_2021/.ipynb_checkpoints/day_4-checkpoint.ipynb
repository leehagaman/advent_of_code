```python
import numpy as np
```

```python
with open("input_day_4.txt") as file:
    lines = file.readlines()
    lines = [line.rstrip() for line in lines]

    
draws = [int(x) for x in lines[0].split(",")]

boards = []
board_markings = []
for b in range((len(lines) - 1) // 6):
    boards.append([])
    board_markings.append([])
    for r in range(5):
        boards[-1].append([int(x) for x in lines[2 + b * 6 + r].split()])
        board_markings[-1].append([0 for x in lines[2 + b * 6 + r].split()])

print(draws)

print("\n\n")

print(boards[0])
print(board_markings[0])
```

```python
for d in draws:
    for i in range(len(boards)):
        # add markings
        for r in range(5):
            for c in range(5):
                if boards[i][r][c] == d:
                    board_markings[i][r][c] = 1
                    
        # check for win
        overall_win = False
        for r in range(5):
            row_win = True
            for c in range(5):
                if board_markings[i][r][c] == 0:
                    row_win = False
            overall_win = max(overall_win, row_win)
        for c in range(5):
            col_win = True
            for r in range(5):
                if board_markings[i][r][c] == 0:
                    col_win = False
            overall_win = max(overall_win, col_win)
            
        if overall_win:
            unmarked_sum = 0
            for r in range(5):
                for c in range(5):
                    if board_markings[i][r][c] == 0:
                        unmarked_sum += boards[i][r][c]
                        
            print(unmarked_sum * d)
            print("done!")
            print(1/0)
            
```

```python
# really should reset board_markings, but it turns out you don't have to 

already_won = [0 for b in boards]

for d in draws:
    for i in range(len(boards)):
        # add markings
        for r in range(5):
            for c in range(5):
                if boards[i][r][c] == d:
                    board_markings[i][r][c] = 1
                    
        # check for win
        overall_win = False
        for r in range(5):
            row_win = True
            for c in range(5):
                if board_markings[i][r][c] == 0:
                    row_win = False
            overall_win = max(overall_win, row_win)
        for c in range(5):
            col_win = True
            for r in range(5):
                if board_markings[i][r][c] == 0:
                    col_win = False
            overall_win = max(overall_win, col_win)
            
        if overall_win and not already_won[i]:
            unmarked_sum = 0
            for r in range(5):
                for c in range(5):
                    if board_markings[i][r][c] == 0:
                        unmarked_sum += boards[i][r][c]
                        
            already_won[i] = 1
                        
            print(unmarked_sum * d)
            print("done!")
            
            
```
