```python
class Expr:
    
    def __init__(self, line=None):
        
        if line == None:
            pass
        
        elif isinstance(line, str):
            
            self.type = "inp"
            self.value = line.zfill(2)
            self.min = 1
            self.max = 9
            
            
        
        elif isinstance(line, int):
            self.type = "int"
            self.value = line
            self.min = line
            self.max = line
        else:
            print("invalid init call")
                
                
    def __repr__(self):
        if self.type == "int":
            return str(self.value)
        elif self.type == "inp":
            return str(self.value)
        elif self.type == "add":
            s = ""
            for arg in self.args:
                if arg.type == "int" or arg.type == "inp":
                    n = str(arg)
                else:
                    n = "(" + str(arg) + ")"
                s += n + " + "
            return s[:-3]
        elif self.type == "mul":
            s = ""
            for arg in self.args:
                if arg.type == "int" or arg.type == "inp":
                    n = str(arg)
                else:
                    n = "(" + str(arg) + ")"
                s += n + " * "
            return s[:-3]
        else:
            
            if self.left.type == "int" or self.left.type == "inp":
                l = str(self.left)
            else:
                l = f"[{self.left}]"
            
            if self.right.type == "int" or self.right.type == "inp":
                r = str(self.right)
            else:
                r = f"[{self.right}]"
            
            if self.type == "mul":
                return f"{l} * {r}"
            elif self.type == "div":
                return f"{l} // {r}"
            elif self.type == "mod":
                return f"{l} % {r}"
            elif self.type == "eql":
                return f"{l} == {r}"
        
    def __add__(self, other, debug=False):
        
        if debug:
            print("adding")
                
        if self.type == "int" and other.type == "int":
            return Expr(self.value + other.value)
            
        elif self.type == "int" and self.value == 0:
            return other
        
        elif other.type == "int" and other.value == 0:
            return self
        
        elif self.type == "add" and other.type == "int": # add new integer to sum
            
            new = Expr()
            new.type = "add"
            new_sum_args = []
        
            added_new_num = False
            for arg in self.args:
                if not added_new_num and arg.type == "int":
                    new_sum_args.append(arg + Expr(other.value))
                    added_new_num = True
                else:
                    new_sum_args.append(arg)
                
            new.args = new_sum_args
            new.min = self.min + other.min
            new.max = self.max + other.max
            
            return new
        
        elif other.type == "add" and self.type == "int": # swapped version of above
            
            new = Expr()
            new.type = "add"
            new_sum_args = []
        
            added_new_num = False
            for arg in other.args:
                if not added_new_num and arg.type == "int":
                    new_sum_args.append(arg + Expr(self.value))
                    added_new_num = True
                else:
                    new_sum_args.append(arg)
                
            new.args = new_sum_args
            new.min = self.min + other.min
            new.max = self.max + other.max
            
            return new
        
        elif self.type == "add" and other.type == "add": # combine ints in the sums
            
            new = Expr()
            new.type = "add"
            new_sum_args = []
        
            int_part = 0
            for arg in self.args + other.args:
                if arg.type == "int":
                    int_part += arg.value
                else:
                    new_sum_args.append(arg)
                
            new.args = new_sum_args + [Expr(int_part)]
            new.min = self.min + other.min
            new.max = self.max + other.max
            
            return new
        
        
        else:
            new = Expr()
            new.type = "add"
            
            if self.type == "add":
                left_list = self.args
            else:
                left_list = [self]
                
            if other.type == "add":
                right_list = other.args
            else:
                right_list = [other]
                
            new.args = left_list + right_list
            new.min = self.min + other.min
            new.max = self.max + other.max
            return new
        
    def __mul__(self, other, debug=False):
        
        if debug:
            print("multiplying")
                
        if self.type == "int" and other.type == "int":
            return Expr(self.value * other.value)
            
        elif self.type == "int" and self.value == 0:
            return Expr(0)
        elif other.type == "int" and other.value == 0:
            return Expr(0)
        
        elif self.type == "int" and self.value == 1:
            return other
        elif other.type == "int" and other.value == 1:
            return self
        
        elif self.type == "mul" and other.type == "int": # add new integer to factorization
            
            #print("debug mul and int")
            #print(self)
            #print(other)
            
            new = Expr()
            new.type = "mul"
            new_sum_args = []
        
            added_new_num = False
            for arg in self.args:
                if not added_new_num and arg.type == "int":
                    new_sum_args.append(arg * Expr(other.value))
                    added_new_num = True
                else:
                    new_sum_args.append(arg)
                
            new.args = new_sum_args
            new.min = np.min([self.min * other.min, self.min * other.max, self.max * other.min, self.max * other.max])
            new.max = np.max([self.min * other.min, self.min * other.max, self.max * other.min, self.max * other.max])
            
            #print(new)
            
            return new
        
        elif other.type == "mul" and self.type == "int": # swapped version of above
            
            new = Expr()
            new.type = "mul"
            new_sum_args = []
        
            added_new_num = False
            for arg in other.args:
                if not added_new_num and arg.type == "int":
                    new_sum_args.append(arg * Expr(self.value))
                    added_new_num = True
                else:
                    new_sum_args.append(arg)
                
            new.args = new_sum_args
            new.min = np.min([self.min * other.min, self.min * other.max, self.max * other.min, self.max * other.max])
            new.max = np.max([self.min * other.min, self.min * other.max, self.max * other.min, self.max * other.max])
            
            return new
        
        elif other.type == "mul" and self.type == "mul": # combine ints in the muls
                    
            new = Expr()
            new.type = "mul"
            new_sum_args = []
        
            int_part = 1
            for arg in self.args + other.args:
                if arg.type == "int":
                    int_part *= arg.value
                else:
                    new_sum_args.append(arg)
                    
            if int_part == 0:
                return Expr(0)
            if int_part == 1:
                new.args = new_sum_args  
            else:
                new.args = new_sum_args + [Expr(int_part)]
            
            new.min = np.min([self.min * other.min, self.min * other.max, self.max * other.min, self.max * other.max])
            new.max = np.max([self.min * other.min, self.min * other.max, self.max * other.min, self.max * other.max])
            
            return new
        
        elif self.type == "add" and other.type == "int": # distribute integer multiplication to the left
            
            new = Expr()
            new.type = "add"
            new_sum_args = []
        
            for arg in self.args:
                new_sum_args.append(arg * Expr(other.value))
                
            new.args = new_sum_args
            new.min = np.min([self.min * other.min, self.min * other.max, self.max * other.min, self.max * other.max])
            new.max = np.max([self.min * other.min, self.min * other.max, self.max * other.min, self.max * other.max])
            
            return new
        
        elif other.type == "add" and self.type == "int": # swapped version of above
            
            new = Expr()
            new.type = "add"
            new_sum_args = []
        
            for arg in other.args:
                new_sum_args.append(arg * Expr(self.value))
                
            new.args = new_sum_args
            new.min = np.min([self.min * other.min, self.min * other.max, self.max * other.min, self.max * other.max])
            new.max = np.max([self.min * other.min, self.min * other.max, self.max * other.min, self.max * other.max])
            
            return new
        
        elif self.type == "add" and other.type == "add": # distribute all addition
            
            #print("debug, double distributing")
            #print("self:", self)
            #print("other:", other)
            
            new = Expr()
            new.type = "add"
            new_sum_args = []
            
            for arg1 in self.args:
                for arg2 in other.args:
                    #new_temp_expr = arg1 * arg2
                    #if new_temp_expr.min < 0:
                        #print("negative min", arg1, arg2)
                    new_sum_args.append(arg1 * arg2)
                    
            new.args = new_sum_args
            
            #for a in new_sum_args:
            #    print(a.min, a.max, ":", a)
            
            new.min = np.sum([x.min for x in new_sum_args])
            new.max = np.sum([x.max for x in new_sum_args])

            #new.min = np.min([self.min * other.min, self.min * other.max, self.max * other.min, self.max * other.max])
            #new.max = np.max([self.min * other.min, self.min * other.max, self.max * other.min, self.max * other.max])
            
            #print("new:", new)
            
            return new
        
        else:
            new = Expr()
            new.type = "mul"
            
            if self.type == "mul":
                left_list = self.args
            else:
                left_list = [self]
                
            if other.type == "mul":
                right_list = other.args
            else:
                right_list = [other]
                
            new.args = left_list + right_list
            
            new.min = np.min([self.min * other.min, self.min * other.max, self.max * other.min, self.max * other.max])
            new.max = np.max([self.min * other.min, self.min * other.max, self.max * other.min, self.max * other.max])
            return new
        
    def __floordiv__(self, other, maybe_distribute=True, debug=False):
        
        if debug:
            print("dividing")
        
            print("floordiv debug")
            print(self, "//", other)
        #print(self.min, self.max, other.min, other.max)
        
        if self.type == "int" and other.type == "int":
            if (self.value > 0 and other.value > 0) or (self.value < 0 and other.value < 0):
                return Expr(self.value // other.value)
            else:
                return Expr(- ((-self.value) // other.value))
            
            
        elif self.type == "int" and self.value == 0:
            return Expr(0)
        elif other.type == "int" and other.value == 1:
            return self
        
        elif other.min == 0 and other.max == 1: # either it's invalid, or it's dividing by one
            return self
        
        if (self.min > 0 or self.max < 0) and (other.min > 0 or other.max < 0): # each is either positive or negative, this could be true but the other case is needed
            self_abs_max = np.max([np.abs(self.max), np.abs(self.min)])
            other_abs_min = np.min([np.abs(other.max), np.abs(other.min)])
            
            if self_abs_max < other_abs_min: # numerator always less than denominator
                return Expr(0)
            
        
            """elif self.min > 0 and other.min > 0 and self.max < other.min: # both positive, other > self, will round to zero
                return Expr(0)

            elif self.max < 0 and other.max < 0 and self.min > other.max: # both negative, other < self, will round to zero
                return Expr(0)"""
                    
        if self.type == "mul" and other.type == "int":
            if self.args[-1].type == "int" and self.args[-1].value % other.value == 0: # exact division
                #print("debug exact floordiv")
                #print("exact self", self)
                #print("exact other", other)
                new = Expr()
                new.type = "mul"
                new_args = self.args[:-1] + [Expr(self.args[-1].value // other.value)]
                
                if self.min > 0 and other.min > 0: # both ranges are positive
            
                    new.min = self.min // other.max
                    new.max = self.max // other.min
                
                elif self.min > 0 and other.max < 0 or self.max < 0 and other.min > 0: # one range is positive, one range is negative
                    new.min = -1 * ((-self.min) // other.max)
                    new.max = -1 * ((-self.max) // other.min)

                else: # at least one range has positive and negative (could restrict this more)
                    new.min = np.min([self.min, self.max, -self.min, -self.max])
                    new.max = np.max([self.min, self.max, -self.min, -self.max])
                
                new_new_args = []
                
                for x in new_args:
                    if str(x) == str(Expr(1)):
                        pass
                    else:
                        new_new_args.append(x)
                        
                
                if len(new_new_args) == 1:
                    return new_new_args[0]
                else:
                    new.args = new_new_args

                    #print("exact result", new)
                    return new
        
        if maybe_distribute and other.type == "int" and self.type == "add": # distribute the //26 to terms of the sum with * 26

            #print("debug distributing floordiv")

            separate_terms = []
            still_modded_terms = []

            for arg in self.args:
                
                should_distribute_this_term = False
                
                if arg.type == "mul":
                    for subarg in arg.args:
                        if subarg.type == "int" and subarg.value % other.value == 0: # there is a multiple of six in this term of the sum
                            should_distribute_this_term = True
                            
                if should_distribute_this_term:
                    separate_terms.append(arg // other)
                else:
                    still_modded_terms.append(arg)

            mod_term = Expr(0)
            for t in still_modded_terms:
                mod_term += t

            separate_term = Expr(0)
            for t in separate_terms:
                separate_term += t

            #print("self", self)
            #print("other", other)
            #print("separate", separate_terms)
            #print("still floor div", still_modded_terms)
            #print("separate min max:", separate_term.min, separate_term.max)
            #for x in separate_term.args:
                #print("    ", x.min, x.max, ":", x)
            #print("mod min max:", mod_term.mi,n, mod_term.max)
            
            #print("result", separate_term + mod_term.__floordiv__(other, maybe_distribute=False))
            
            return separate_term + mod_term.__floordiv__(other, maybe_distribute=False)

        
        else:
            new = Expr()
            new.type = "div"
            new.left = self
            new.right = other
                        
            if self.min > 0 and other.min > 0: # both ranges are positive
            
                new.min = self.min // other.max
                new.max = self.max // other.min
                
            elif self.min > 0 and other.max < 0 or self.max < 0 and other.min > 0: # one range is positive, one range is negative
                new.min = -1 * ((-self.min) // other.max)
                new.max = -1 * ((-self.max) // other.min)
                
            else: # at least one range has positive and negative (could restrict this more)
                new.min = np.min([self.min, self.max, -self.min, -self.max])
                new.max = np.max([self.min, self.max, -self.min, -self.max])
            
            return new
        
        
    def __mod__(self, other, already_checked=False, debug=False): # specifies that these are never negative
        
        if debug:
            print("modding")
                        
        if self.type == "int" and other.type == "int":
            return Expr(self.value % other.value)
            
        elif self.type == "int" and self.value == 0:
            return Expr(0)
        elif other.type == "int" and other.value == 1:
            return Expr(0)
        
        elif str(self) == str(other):
            return Expr(0)
                
        elif self.type == "int" and other.type == "int" and self.min > 0 and self.max > 0 and self.max < other.min:
            return self
        
        elif self.type == "mul" and other.type == "int":
            for arg in self.args:
                if str(arg.__mod__(other, debug=False)) == str(Expr(0)):
                    return Expr(0)
                
            new = Expr()
            new.type = "mod"
            new.left = self
            new.right = other
            new.min = 0
            new.max = other.max - 1 # could restrict this more if the range of self is all positive
            
            return new
        
        elif self.type == "add" and already_checked == False: # throw away terms that are 0 in the mod
            
            """if debug:
            
                print("\nmod debug", debug)

                print("self", self)
                print("other", other)"""

            new = Expr()
            new.type = "add"

            new_sum_args = []
            """if debug:
                print("checking for zero:")"""
            for arg in self.args:
                #print(arg, ", is mod zero: ", str(arg % other) == str(Expr(0)))
                if str(arg.__mod__(other, debug=False)) == str(Expr(0)):
                    #print("mod is zero", arg)
                    pass
                else:
                    new_sum_args.append(arg)
                    
            """if debug:
                print("done checking for zero")"""

            if len(new_sum_args) == 0:
                """if debug:
                    print("result", Expr(0))"""
                return Expr(0)
            else:

                new.args = new_sum_args
                new.min = 0
                new.max = np.min([other.max - 1, self.max])

                """if debug:
                    print("result", new.__mod__(other, already_checked=True))"""

                return new.__mod__(other, already_checked=True) # still need the mod here, didn't distribute fully
        
        else:
            new = Expr()
            new.type = "mod"
            new.left = self
            new.right = other
            new.min = 0
            new.max = np.min([other.max - 1, self.max])
            
            return new
        
    def __eq__(self, other, debug=False):
        
        if debug:
            print("equaling", self, other)
        
        if self.type == "int" and other.type == "int":
            return Expr(int(self.value == other.value))
        
        elif str(self) == str(other):
            return Expr(1)
        
        elif self.min == self.max and other.min == other.max and self.min == other.min:
            return Expr(1)
        
        elif self.max < other.min or other.max < self.min:
            return Expr(0)
        
        else:
            new = Expr()
            new.type = "eql"
            new.left = self
            new.right = other
            new.min = 0
            new.max = 1
            
            #print("nothing special", new)
            
            """if str(new) == "[[[inp3 + 5748] % 26] == inp4] == 0":
                return Expr(0) # also try one
            
            if str(new) == "[[([inp7 + 7553] % 26) + -8] == inp8] == 0":
                return Expr(0)
            
            if str(new) == "[[([inp6 + ([inp7 + 7553] // 26)] % 26) + -9] == inp9] == 0":
                return Expr(0)"""
            
            
            return new
        
    def refine(self): # if somehow the min and max get restricted to a single value, this will help to replace it with an int
        if self.min == self.max and self.type != "int":
            return Expr(self.min)
        else:
            return self
```
