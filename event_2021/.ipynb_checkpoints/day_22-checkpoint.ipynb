```python
def num_cubes_on(all_cuboids):
    tot = 0
    
    for a in all_cuboids:
    
        a_xlims = a[0]
        a_ylims = a[1]
        a_zlims = a[2]
        
        tot += ((a_xlims[1] - a_xlims[0] + 1) * 
                (a_ylims[1] - a_ylims[0] + 1) * 
                (a_zlims[1] - a_zlims[0] + 1))
        
    return tot

def subtract(a, b, debug=False): 
        
    a_xlims = a[0]
    a_ylims = a[1]
    a_zlims = a[2]
    
    b_xlims = b[0]
    b_ylims = b[1]
    b_zlims = b[2]
    
    if (a_xlims[1] < b_xlims[0] or b_xlims[1] < a_xlims[0] or
        a_ylims[1] < b_ylims[0] or b_ylims[1] < a_ylims[0] or
        a_zlims[1] < b_zlims[0] or b_zlims[1] < a_zlims[0]): # x or y or z overlaps are totally independent, no intersection
        return [a]
    
    x_min = min(a[0][0], b[0][0])
    y_min = min(a[1][0], b[1][0])
    z_min = min(a[2][0], b[2][0])
    
    x_max = max(a[0][1], b[0][1])
    y_max = max(a[1][1], b[1][1])
    z_max = max(a[2][1], b[2][1])
    
    if a_xlims[0] == b_xlims[0] and a_xlims[1] == b_xlims[1]: # same range, no splits neede
        x_split_pairs = [[a_xlims[0], a_xlims[1]]]
        
    elif a_xlims[0] == b_xlims[0] and a_xlims[0] == b_xlims[1]: # three shared on left
        x_split_pairs = [[a_xlims[0], a_xlims[0]], [a_xlims[0] + 1, a_xlims[1]]]
    elif a_xlims[0] == b_xlims[0] and a_xlims[0] == a_xlims[1]: # three shared on left
        x_split_pairs = [[a_xlims[0], a_xlims[0]], [a_xlims[0] + 1, b_xlims[1]]]
        
    elif a_xlims[1] == b_xlims[1] and a_xlims[1] == b_xlims[0]: # three shared on right
        x_split_pairs = [[a_xlims[0], a_xlims[1] - 1], [a_xlims[1], a_xlims[1]]]
    elif a_xlims[1] == b_xlims[1] and a_xlims[1] == a_xlims[0]: # three shared on right
        x_split_pairs = [[b_xlims[0], a_xlims[1] - 1], [a_xlims[1], a_xlims[1]]]
        
    elif b_xlims[0] < a_xlims[0] and a_xlims[1] < b_xlims[1]: # a contained within b
        x_split_pairs = [[b_xlims[0], a_xlims[0] - 1], [a_xlims[0], a_xlims[1]], [a_xlims[1] + 1, b_xlims[1]]]
    elif a_xlims[0] < b_xlims[0] and b_xlims[1] < a_xlims[1]: # b contained within a
        x_split_pairs = [[a_xlims[0], b_xlims[0] - 1], [b_xlims[0], b_xlims[1]], [b_xlims[1] + 1, a_xlims[1]]]
        
    elif a_xlims[0] < b_xlims[0] and a_xlims[1] < b_xlims[1]: # partially overlap, a left of b
        x_split_pairs = [[a_xlims[0], b_xlims[0] - 1], [b_xlims[0], a_xlims[1]], [a_xlims[1] + 1, b_xlims[1]]]
    elif b_xlims[0] < a_xlims[0] and b_xlims[1] < a_xlims[1]: # partially overlap, b left of a
        x_split_pairs = [[b_xlims[0], a_xlims[0] - 1], [a_xlims[0], b_xlims[1]], [b_xlims[1] + 1, a_xlims[1]]]
        
    elif a_xlims[0] == b_xlims[0] and a_xlims[1] < b_xlims[1]: # left shared, right has a < b
        x_split_pairs = [[a_xlims[0], a_xlims[1]], [a_xlims[1] + 1, b_xlims[1]]]
    elif a_xlims[0] == b_xlims[0] and a_xlims[1] > b_xlims[1]: # left shared, right has a > b
        x_split_pairs = [[b_xlims[0], b_xlims[1]], [b_xlims[1] + 1, a_xlims[1]]]
        
    elif a_xlims[1] == b_xlims[1] and a_xlims[0] < b_xlims[0]: # right shared, left has a < b
        x_split_pairs = [[a_xlims[0], b_xlims[0] - 1], [b_xlims[0], b_xlims[1]]]
    elif a_xlims[1] == b_xlims[1] and a_xlims[0] > b_xlims[0]: # right shared, left has a > b
        x_split_pairs = [[b_xlims[0], a_xlims[0] - 1], [a_xlims[0], b_xlims[1]]]
        
    else:
        print("invalid ranges!")
        
    if a_ylims[0] == b_ylims[0] and a_ylims[1] == b_ylims[1]: # same range, no splits neede
        y_split_pairs = [[a_ylims[0], a_ylims[1]]]   
    elif a_ylims[0] == b_ylims[0] and a_ylims[0] == b_ylims[1]: # three shared on left
        y_split_pairs = [[a_ylims[0], a_ylims[0]], [a_ylims[0] + 1, a_ylims[1]]]
    elif a_ylims[0] == b_ylims[0] and a_ylims[0] == a_ylims[1]: # three shared on left
        y_split_pairs = [[a_ylims[0], a_ylims[0]], [a_ylims[0] + 1, b_ylims[1]]]    
    elif a_ylims[1] == b_ylims[1] and a_ylims[1] == b_ylims[0]: # three shared on right
        y_split_pairs = [[a_ylims[0], a_ylims[1] - 1], [a_ylims[1], a_ylims[1]]]
    elif a_ylims[1] == b_ylims[1] and a_ylims[1] == a_ylims[0]: # three shared on right
        y_split_pairs = [[b_ylims[0], a_ylims[1] - 1], [a_ylims[1], a_ylims[1]]]
    elif b_ylims[0] < a_ylims[0] and a_ylims[1] < b_ylims[1]: # a contained within b
        y_split_pairs = [[b_ylims[0], a_ylims[0] - 1], [a_ylims[0], a_ylims[1]], [a_ylims[1] + 1, b_ylims[1]]]
    elif a_ylims[0] < b_ylims[0] and b_ylims[1] < a_ylims[1]: # b contained within a
        y_split_pairs = [[a_ylims[0], b_ylims[0] - 1], [b_ylims[0], b_ylims[1]], [b_ylims[1] + 1, a_ylims[1]]]    
    elif a_ylims[0] < b_ylims[0] and a_ylims[1] < b_ylims[1]: # partially overlap, a left of b
        y_split_pairs = [[a_ylims[0], b_ylims[0] - 1], [b_ylims[0], a_ylims[1]], [a_ylims[1] + 1, b_ylims[1]]]
    elif b_ylims[0] < a_ylims[0] and b_ylims[1] < a_ylims[1]: # partially overlap, b left of a
        y_split_pairs = [[b_ylims[0], a_ylims[0] - 1], [a_ylims[0], b_ylims[1]], [b_ylims[1] + 1, a_ylims[1]]]    
    elif a_ylims[0] == b_ylims[0] and a_ylims[1] < b_ylims[1]: # left shared, right has a < b
        y_split_pairs = [[a_ylims[0], a_ylims[1]], [a_ylims[1] + 1, b_ylims[1]]]
    elif a_ylims[0] == b_ylims[0] and a_ylims[1] > b_ylims[1]: # left shared, right has a > b
        y_split_pairs = [[b_ylims[0], b_ylims[1]], [b_ylims[1] + 1, a_ylims[1]]]    
    elif a_ylims[1] == b_ylims[1] and a_ylims[0] < b_ylims[0]: # right shared, left has a < b
        y_split_pairs = [[a_ylims[0], b_ylims[0] - 1], [b_ylims[0], b_ylims[1]]]
    elif a_ylims[1] == b_ylims[1] and a_ylims[0] > b_ylims[0]: # right shared, left has a > b
        y_split_pairs = [[b_ylims[0], a_ylims[0] - 1], [a_ylims[0], b_ylims[1]]]
    else:
        print("invalid ranges!")
        
    if a_zlims[0] == b_zlims[0] and a_zlims[1] == b_zlims[1]: # same range, no splits neede
        z_split_pairs = [[a_zlims[0], a_zlims[1]]]   
    elif a_zlims[0] == b_zlims[0] and a_zlims[0] == b_zlims[1]: # three shared on left
        z_split_pairs = [[a_zlims[0], a_zlims[0]], [a_zlims[0] + 1, a_zlims[1]]]
    elif a_zlims[0] == b_zlims[0] and a_zlims[0] == a_zlims[1]: # three shared on left
        z_split_pairs = [[a_zlims[0], a_zlims[0]], [a_zlims[0] + 1, b_zlims[1]]]    
    elif a_zlims[1] == b_zlims[1] and a_zlims[1] == b_zlims[0]: # three shared on right
        z_split_pairs = [[a_zlims[0], a_zlims[1] - 1], [a_zlims[1], a_zlims[1]]]
    elif a_zlims[1] == b_zlims[1] and a_zlims[1] == a_zlims[0]: # three shared on right
        z_split_pairs = [[b_zlims[0], a_zlims[1] - 1], [a_zlims[1], a_zlims[1]]]
    elif b_zlims[0] < a_zlims[0] and a_zlims[1] < b_zlims[1]: # a contained within b
        z_split_pairs = [[b_zlims[0], a_zlims[0] - 1], [a_zlims[0], a_zlims[1]], [a_zlims[1] + 1, b_zlims[1]]]
    elif a_zlims[0] < b_zlims[0] and b_zlims[1] < a_zlims[1]: # b contained within a
        z_split_pairs = [[a_zlims[0], b_zlims[0] - 1], [b_zlims[0], b_zlims[1]], [b_zlims[1] + 1, a_zlims[1]]]    
    elif a_zlims[0] < b_zlims[0] and a_zlims[1] < b_zlims[1]: # partially overlap, a left of b
        z_split_pairs = [[a_zlims[0], b_zlims[0] - 1], [b_zlims[0], a_zlims[1]], [a_zlims[1] + 1, b_zlims[1]]]
    elif b_zlims[0] < a_zlims[0] and b_zlims[1] < a_zlims[1]: # partially overlap, b left of a
        z_split_pairs = [[b_zlims[0], a_zlims[0] - 1], [a_zlims[0], b_zlims[1]], [b_zlims[1] + 1, a_zlims[1]]]    
    elif a_zlims[0] == b_zlims[0] and a_zlims[1] < b_zlims[1]: # left shared, right has a < b
        z_split_pairs = [[a_zlims[0], a_zlims[1]], [a_zlims[1] + 1, b_zlims[1]]]
    elif a_zlims[0] == b_zlims[0] and a_zlims[1] > b_zlims[1]: # left shared, right has a > b
        z_split_pairs = [[b_zlims[0], b_zlims[1]], [b_zlims[1] + 1, a_zlims[1]]]    
    elif a_zlims[1] == b_zlims[1] and a_zlims[0] < b_zlims[0]: # right shared, left has a < b
        z_split_pairs = [[a_zlims[0], b_zlims[0] - 1], [b_zlims[0], b_zlims[1]]]
    elif a_zlims[1] == b_zlims[1] and a_zlims[0] > b_zlims[0]: # right shared, left has a > b
        z_split_pairs = [[b_zlims[0], a_zlims[0] - 1], [a_zlims[0], b_zlims[1]]]
    else:
        print("invalid ranges!")
        
        
    """x_count_change_pos = []
    for i in range(x_max - x_min + 1):
        left = x_min + i - 1
        center = x_min + i
        right = x_min + i + 1
        newly_added = []
        #print("considering", center)
        if i == 0 or i == x_max - x_min:
            x_count_change_pos.append(center) # starts new block
            newly_added.append(center)
            #print("newly added", newly_added)
            continue
        num_on_left = 0
        if a_xlims[0] <= left <= a_xlims[1]:
            num_on_left += 1
        if b_xlims[0] <= left <= b_xlims[1]:
            num_on_left += 1
        num_on_right = 0
        if a_xlims[0] <= right <= a_xlims[1]:
            num_on_right += 1
        if b_xlims[0] <= right <= b_xlims[1]:
            num_on_right += 1
        num_in_center = 0
        if a_xlims[0] <= center <= a_xlims[1]:
            num_in_center += 1
        if b_xlims[0] <= center <= b_xlims[1]:
            num_in_center += 1            
        if (num_on_left < num_in_center):
            x_count_change_pos.append(left)
            x_count_change_pos.append(center)
            newly_added.append(left)
            newly_added.append(center)
        if (num_in_center > num_on_right):
            x_count_change_pos.append(center)
            x_count_change_pos.append(right)
            newly_added.append(center)
            newly_added.append(right)
        #print("left center right", left, center, right)
        #print("left center right counts", num_on_left, num_in_center, num_on_right)
        #print("newly added", newly_added)
    if len(x_count_change_pos) == 1: # start and end the same, just one pair
        x_split_pairs = [[x_count_change_pos[0], x_count_change_pos[0]]] # make the start and end the same to give one pair
    elif len(x_count_change_pos) == 2: # start and end are the only things, make two pairs
        x_split_pairs = [[x_count_change_pos[0], x_count_change_pos[0]], [x_count_change_pos[1], x_count_change_pos[1]]]
    else:
        x_split_pairs = []         
        for i in range(len(x_count_change_pos)):
            if i % 2 == 0:
                x_split_pairs.append([x_count_change_pos[i]])
            else:
                x_split_pairs[-1].append(x_count_change_pos[i])"""
            
    if debug:
        print("intersecting with", a)
        print("splits")
        print(f"x_split_pairs", x_split_pairs)
        print(f"y_split_pairs", y_split_pairs)
        print(f"z_split_pairs", z_split_pairs)

    for pairs in [x_split_pairs, y_split_pairs, z_split_pairs]:
        for pair in pairs:
            if pair[0] > pair[1]:
                print("big problem, pair in wrong order!", pair)

        for i in range(len(pairs) - 1):
            if pairs[i+1][0] - pairs[i][1] != 1:
                print("big problem, gap in pairs!", pairs[i], pairs[i+1])

    ret = []
    
    for i_x in range(len(x_split_pairs)):
        for i_y in range(len(y_split_pairs)):
            for i_z in range(len(z_split_pairs)):
                
                new_cuboid = [x_split_pairs[i_x], y_split_pairs[i_y], z_split_pairs[i_z]]
                    
                in_a = (a[0][0] <= new_cuboid[0][0] <= a[0][1] and a[0][0] <= new_cuboid[0][1] <= a[0][1] and
                         a[1][0] <= new_cuboid[1][0] <= a[1][1] and a[1][0] <= new_cuboid[1][1] <= a[1][1] and
                         a[2][0] <= new_cuboid[2][0] <= a[2][1] and a[2][0] <= new_cuboid[2][1] <= a[2][1])
                
                in_b = (b[0][0] <= new_cuboid[0][0] <= b[0][1] and b[0][0] <= new_cuboid[0][1] <= b[0][1] and
                         b[1][0] <= new_cuboid[1][0] <= b[1][1] and b[1][0] <= new_cuboid[1][1] <= b[1][1] and
                         b[2][0] <= new_cuboid[2][0] <= b[2][1] and b[2][0] <= new_cuboid[2][1] <= b[2][1])
                
                
                if debug:
                    print(new_cuboid, in_a, in_b)
                    
                    if in_a and not in_b:
                        print("added")
                    else:
                        print("skipped")


                if in_a and not in_b:
                    ret.append(new_cuboid)
                    
    if debug:
        print("subtraction result: ", ret)
                      
    return ret
    
    
    
        
```

```python
# general idea:
# have a list of cuboids that are turned on, all non-overlapping
# when adding a new cube, intersect it with each existing cuboid
# each intersection will return the new cuboids that should be turned on from that intersection. 

# could have two intersections each non-overlapping, in which case the entire new region will be added twice. 
# solve for this by requiring that the output isn't already fully contained in one of the new cuboids.
```

```python
all_cuboids = [[x_lims[0], y_lims[0], z_lims[0]]]


# example:
# 27
# 27 + 19 = 46
# 27 + 19 - 8 = 38
# 39

#print(lines[0])

show_counts = False
show_cuboids = False
show_commands = False
debug = False


for i in tqdm(range(1, len(on_off_list))): # iterating over lines
    
    if show_counts:
        print(f"\nnum cuboids: {len(all_cuboids)}")
        print(f"num cubes on: {num_cubes_on(all_cuboids)}")
        
    if show_cuboids:
        for x in all_cuboids:
            print(x, num_cubes_on([x]))
            
    if show_commands:
        print("\n")
        print(lines[i])

    
    #if x_lims[i][0] < -1000: # stop for part 1
    #    break
        
    new_all_cuboids = []

    for c in all_cuboids: # iterating over cuboids in the previous step
        new_all_cuboids += subtract(c, [x_lims[i], y_lims[i], z_lims[i]], debug=debug)

    if on_off_list[i]:
        new_all_cuboids.append([x_lims[i], y_lims[i], z_lims[i]])
        
    all_cuboids = new_all_cuboids

print(f"\nnum cuboids: {len(all_cuboids)}")
print(f"num cubes on: {num_cubes_on(all_cuboids)}")
if show_cuboids:
    for x in all_cuboids:
        print(x, num_cubes_on([x]))
```
