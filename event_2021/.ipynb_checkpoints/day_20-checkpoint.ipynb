```python
def expand_with_zeros(strs, expand_num):

    row_of_dots = ""
    for i in range(len(strs[0]) + 2 * expand_num):
        row_of_dots += "."

    ten_dots = ""
    for i in range(expand_num):
        ten_dots += "."

    new_strs = []
    for i in range(expand_num):
        new_strs.append(row_of_dots)

    for r in range(len(strs)):
        new_strs.append(ten_dots)
        for c in range(len(strs[0])):
            new_strs[-1] += strs[r][c]

        new_strs[-1] += ten_dots

    for i in range(expand_num):
        new_strs.append(row_of_dots)
        
    return new_strs

def expand_with_ones(strs, expand_num):

    row_of_dots = ""
    for i in range(len(strs[0]) + 2 * expand_num):
        row_of_dots += "#"

    ten_dots = ""
    for i in range(expand_num):
        ten_dots += "#"

    new_strs = []
    for i in range(expand_num):
        new_strs.append(row_of_dots)

    for r in range(len(strs)):
        new_strs.append(ten_dots)
        for c in range(len(strs[0])):
            new_strs[-1] += strs[r][c]

        new_strs[-1] += ten_dots

    for i in range(expand_num):
        new_strs.append(row_of_dots)
        
    return new_strs

def crop(strs, crop_num):
    new_strs = []
    for r in range(crop_num, len(strs) - crop_num):
        new_strs.append("")
        for c in range(crop_num, len(strs[0]) - crop_num):
            new_strs[-1] += strs[r][c]
            
    return new_strs
```

```python
def transform(strs, outside_bits):

    output_strs = []
    
    #print(len(strs), len(strs[0]))

    for r in range(len(strs)):
        output_strs.append("")
        for c in range(len(strs[0])):
            n = 0
            for i in range(9):
                dr, dc = [[r - 1, c - 1],
                               [r - 1, c],
                               [r - 1, c + 1],
                               [r, c - 1],
                               [r, c],
                               [r, c + 1],
                               [r + 1, c - 1],
                               [r + 1, c],
                               [r + 1, c + 1]][i]

                if dr < 0 or dc < 0 or dr > len(strs) - 1 or dc > len(strs[0]) - 1:
                    if outside_bits == 1:
                        n += 2**(8 - i)
                elif strs[dr][dc] == ".":
                    pass
                elif strs[dr][dc] == "#":
                    #print("adding to n")
                    n += 2**(8 - i)
                else:
                    print("bad! 1")
            
            #print(n)
            if alg[n] == "#":
                output_strs[-1] += "#"
            elif alg[n] == ".":
                output_strs[-1] += "."
            else:
                print("bad! 2")
        
    return output_strs

og = expand_with_zeros(strs, 1)

res1 = expand_with_ones(transform(og, 0), 1)

res2 = expand_with_zeros(transform(res1, 1), 1)

#cropped_res2 = crop(res2, 7)

num_lit_0 = 0
for r in range(len(strs)):
    for c in range(len(strs[0])):
        if strs[r][c] == "#":
            num_lit_0 += 1

num_lit_1 = 0
for r in range(len(res1)):
    for c in range(len(res1[0])):
        if res1[r][c] == "#":
            num_lit_1 += 1
            
num_lit_2 = 0
for r in range(len(res2)):
    for c in range(len(res2[0])):
        if res2[r][c] == "#":
            num_lit_2 += 1
            
"""num_lit_crop_2 = 0
for r in range(len(cropped_res2)):
    for c in range(len(cropped_res2[0])):
        if cropped_res2[r][c] == "#":
            num_lit_crop_2 += 1"""
            
            
print(num_lit_0)

for r in strs:
    print(r)
            
print(num_lit_1)        
        
for r in res1:
    print(r)
    
print(num_lit_2)        
        
for r in res2:
    print(r)
    
"""print(num_lit_crop_2)        
        
for r in cropped_res2:
    print(r)"""
```
