```python
possible_total_costs = []

existing_states = []

def finished_state(state):
    hall = state[0]
    row1 = state[1]
    row2 = state[2]
    total_cost_so_far = state[3]
    num_moves = state[4]

    return row1 == ["A", "B", "C", "D"] and row2 == ["A", "B", "C", "D"]

def print_state(state):
    
    hall = state[0]
    row1 = state[1]
    row2 = state[2]
    total_cost_so_far = state[3]
    num_moves = state[4]
        
    print("\n#############", end="")
    print("     total cost so far:", total_cost_so_far)
    print("#", end="")
    for h in hall:
        print(h, end="")
    print("#", end="")
    print("     num moves so far:", num_moves)
    print(f"###{row1[0]}#{row1[1]}#{row1[2]}#{row1[3]}###")
    print(f"  #{row2[0]}#{row2[1]}#{row2[2]}#{row2[3]}#  ")
    print("  #########")

def make_moves(state, debug=False):
    
    global existing_states
        
    hall = state[0]
    row1 = state[1]
    row2 = state[2]
    total_cost_so_far = state[3]
    num_moves = state[4]

    for place_int in range(3):
        place = [hall, row1, row2][place_int]
        for i in range(len(place)):
            
            if debug:
                print(f"\ntrying to move the {i}th position of ", end="")
                if place_int == 0:
                    print("the hall")
                elif place_int == 1:
                    print("row 1")
                elif place_int == 2:
                    print("row 2")
        
            c = place[i]
            if c == ".": # no pod to move
                if debug:
                    print("no pod there")
                continue


            if place == hall: # move pod into final room

                if c == "A":
                    obstructed = False
                    hall_lower = min(i, 2)
                    hall_upper = max(i, 2)
                    for j in range(hall_lower, hall_upper + 1):
                        if j == i: # existing position, skip
                            continue
                        elif hall[j] != ".": # there's an obstruction
                            obstructed = True
                    if obstructed:
                        if debug:
                            print("path obstructed")
                        continue
                    if row1[0] == "." and row2[0] == ".": # move into row2
                        new_hall = hall[:i] + ["."] + hall[i+1:]
                        new_row2 = row2[:0] + ["A"] + row2[1:]
                        new_total_cost_so_far = total_cost_so_far + 1 * abs(i - 2) + 1 * 2
                        existing_states.append((new_hall, row1, new_row2, new_total_cost_so_far, num_moves + 1))
                        if debug:
                            print("moved it")
                        continue
                    elif row1[0] == "." and row2[0] == "A": # move into row1
                        new_hall = hall[:i] + ["."] + hall[i+1:]
                        new_row1 = row1[:0] + ["A"] + row1[1:]
                        new_total_cost_so_far = total_cost_so_far + 1 * abs(i - 2) + 1 * 1
                        existing_states.append((new_hall, new_row1, row2, new_total_cost_so_far, num_moves + 1))
                        if debug:
                            print("moved it")
                        continue
                    if debug:
                        print("other pod in correct spot")
                        continue
                elif c == "B":
                    obstructed = False
                    hall_lower = min(i, 4)
                    hall_upper = max(i, 4)
                    for j in range(hall_lower, hall_upper + 1):
                        if j == i: # existing position, skip
                            continue
                        elif hall[j] != ".": # there's an obstruction
                            obstructed = True
                    if obstructed:
                        if debug:
                            print("path obstructed")
                        continue
                    if row1[1] == "." and row2[1] == ".": # move into row2
                        new_hall = hall[:i] + ["."] + hall[i+1:]
                        new_row2 = row2[:1] + ["B"] + row2[2:]
                        new_total_cost_so_far = total_cost_so_far + 10 * abs(i - 4) + 10 * 2
                        existing_states.append((new_hall, row1, new_row2, new_total_cost_so_far, num_moves + 1))
                        if debug:
                            print("moved it")
                        continue
                    elif row1[1] == "." and row2[1] == "B": # move into row1
                        new_hall = hall[:i] + ["."] + hall[i+1:]
                        new_row1 = row1[:1] + ["B"] + row1[2:]
                        new_total_cost_so_far = total_cost_so_far + 10 * abs(i - 4) + 10 * 1
                        existing_states.append((new_hall, new_row1, row2, new_total_cost_so_far, num_moves + 1))
                        if debug:
                            print("moved it")
                        continue
                    if debug:
                        print("other pod in correct spot")
                        continue
                elif c == "C":
                    obstructed = False
                    hall_lower = min(i, 6)
                    hall_upper = max(i, 6)
                    for j in range(hall_lower, hall_upper + 1):
                        if j == i: # existing position, skip
                            continue
                        elif hall[j] != ".": # there's an obstruction
                            obstructed = True
                    if obstructed:
                        if debug:
                            print("path obstructed")
                        continue
                    if row1[2] == "." and row2[2] == ".": # move into row2
                        new_hall = hall[:i] + ["."] + hall[i+1:]
                        new_row2 = row2[:2] + ["C"] + row2[3:]
                        new_total_cost_so_far = total_cost_so_far + 100 * abs(i - 6) + 100 * 2
                        existing_states.append((new_hall, row1, new_row2, new_total_cost_so_far, num_moves + 1))
                        if debug:
                            print("moved it")
                        continue
                    elif row1[2] == "." and row2[2] == "C": # move into row1
                        new_hall = hall[:i] + ["."] + hall[i+1:]
                        new_row1 = row1[:2] + ["C"] + row1[3:]
                        new_total_cost_so_far = total_cost_so_far + 100 * abs(i - 6) + 100 * 1
                        existing_states.append((new_hall, new_row1, row2, new_total_cost_so_far, num_moves + 1))
                        if debug:
                            print("moved it")
                        continue
                    if debug:
                        print("other pod in correct spot")
                        continue
                elif c == "D":

                    obstructed = False
                    hall_lower = min(i, 8)
                    hall_upper = max(i, 8)
                    for j in range(hall_lower, hall_upper + 1):
                        if j == i: # existing position, skip
                            continue
                        elif hall[j] != ".": # there's an obstruction
                            obstructed = True
                    if obstructed:
                        if debug:
                            print("path obstructed")
                        continue

                    if row1[3] == "." and row2[3] == ".": # move into row2
                        new_hall = hall[:i] + ["."] + hall[i+1:]
                        new_row2 = row2[:3] + ["D"] + row2[4:]
                        new_total_cost_so_far = total_cost_so_far + 1000 * abs(i - 8) + 1000 * 2
                        existing_states.append((new_hall, row1, new_row2, new_total_cost_so_far, num_moves + 1))
                        if debug:
                            print("moved it")
                        continue
                    elif row1[3] == "." and row2[3] == "D": # move into row1
                        new_hall = hall[:i] + ["."] + hall[i+1:]
                        new_row1 = row1[:3] + ["D"] + row1[4:]
                        new_total_cost_so_far = total_cost_so_far + 1000 * abs(i - 8) + 1000 * 1
                        existing_states.append((new_hall, new_row1, row2, new_total_cost_so_far, num_moves + 1))
                        if debug:
                            print("moved it")
                        continue
                    if debug:
                        print("other pod in correct spot")
                        continue

            else: # moving from one of the rows
                
                for hall_pos in range(11):

                    if debug:
                        print("trying to move it to", hall_pos)
                    if hall_pos == 2 or hall_pos == 4 or hall_pos == 6 or hall_pos == 8 or hall[hall_pos] != ".":
                        if debug:
                            print("illegal/occupied hallway stopping point")
                        continue
                    obstructed = False
                    hall_lower = min(hall_pos, i * 2 + 2)
                    hall_upper = max(hall_pos, i * 2 + 2)
                    if debug:
                        print(f"checking for obstructions between {hall_lower} and {hall_upper} (inclusive)")
                    for j in range(hall_lower, hall_upper + 1):
                        if j == hall_pos: # existing pod, continue
                            continue
                        elif hall[j] != ".": # there's an obstruction
                            obstructed = True
                    if obstructed:
                        if debug:
                            print("path obstructed")
                        continue

                    if c == "A":
                        new_hall = hall[:hall_pos] + ["A"] + hall[hall_pos+1:]
                    elif c == "B":
                        new_hall = hall[:hall_pos] + ["B"] + hall[hall_pos+1:]
                    elif c == "C":
                        new_hall = hall[:hall_pos] + ["C"] + hall[hall_pos+1:]
                    elif c == "D":
                        new_hall = hall[:hall_pos] + ["D"] + hall[hall_pos+1:]

                    ordered_new_hall_occupants = []

                    for h in new_hall:
                        if h != ".":
                            ordered_new_hall_occupants.append(h)

                    for p in range(len(ordered_new_hall_occupants) - 1):
                        if ordered_new_hall_occupants[p] > ordered_new_hall_occupants[p + 1]: # out of order
                            if debug:
                                print("would make hallway out of order")
                            continue


                    if place == row1:

                        if i == 0 and row1[i] == "A" and row2[i] == "A": # already placed, don't move
                            if debug:
                                print("already placed correctly")
                            continue
                        elif i == 1 and row1[i] == "B" and row2[i] == "B": # already placed, don't move
                            if debug:
                                print("already placed correctly")
                            continue
                        elif i == 2 and row1[i] == "C" and row2[i] == "C": # already placed, don't move
                            if debug:
                                print("already placed correctly")
                            continue
                        elif i == 3 and row1[i] == "D" and row2[i] == "D": # already placed, don't move
                            if debug:
                                print("already placed correctly")
                            continue

                        if c == "A":
                            new_row1 = row1[:i] + ["."] + row1[i+1:]
                            new_total_cost_so_far = total_cost_so_far + 1 * abs([2, 4, 6, 8][i] - hall_pos) + 1 * 1
                            existing_states.append((new_hall, new_row1, row2, new_total_cost_so_far, num_moves + 1))
                            if debug:
                                print("moved it")
                            continue
                        elif c == "B":
                            new_row1 = row1[:i] + ["."] + row1[i+1:]
                            new_total_cost_so_far = total_cost_so_far + 10 * abs([2, 4, 6, 8][i] - hall_pos) + 10 * 1
                            existing_states.append((new_hall, new_row1, row2, new_total_cost_so_far, num_moves + 1))
                            if debug:
                                print("moved it")
                            continue
                        elif c == "C":
                            new_row1 = row1[:i] + ["."] + row1[i+1:]
                            new_total_cost_so_far = total_cost_so_far + 100 * abs([2, 4, 6, 8][i] - hall_pos) + 100 * 1
                            existing_states.append((new_hall, new_row1, row2, new_total_cost_so_far, num_moves + 1))
                            if debug:
                                print("moved it")
                            continue
                        elif c == "D":
                            new_row1 = row1[:i] + ["."] + row1[i+1:]
                            new_total_cost_so_far = total_cost_so_far + 1000 * abs([2, 4, 6, 8][i] - hall_pos) + 1000 * 1
                            existing_states.append((new_hall, new_row1, row2, new_total_cost_so_far, num_moves + 1))
                            if debug:
                                print("moved it")
                            continue

                    elif place == row2:

                        if i == 0 and row2[i] == "A": # already placed, don't move
                            if debug:
                                print("already placed correctly")
                            continue
                        elif i == 1 and row2[i] == "B": # already placed, don't move
                            if debug:
                                print("already placed correctly")
                            continue
                        elif i == 2 and row2[i] == "C": # already placed, don't move
                            if debug:
                                print("already placed correctly")
                            continue
                        elif i == 3 and row2[i] == "D": # already placed, don't move
                            if debug:
                                print("already placed correctly")
                            continue

                        if row1[i] != ".": # blocked by row1
                            if debug:
                                print("blocked by row 1")
                            continue

                        if c == "A":
                            new_row2 = row2[:i] + ["."] + row2[i+1:]
                            new_total_cost_so_far = total_cost_so_far + 1 * abs([2, 4, 6, 8][i] - hall_pos) + 1 * 2
                            existing_states.append((new_hall, row1, new_row2, new_total_cost_so_far, num_moves + 1))
                            if debug:
                                print("moved it")
                            continue
                        elif c == "B":
                            new_row2 = row2[:i] + ["."] + row2[i+1:]
                            new_total_cost_so_far = total_cost_so_far + 10 * abs([2, 4, 6, 8][i] - hall_pos) + 10 * 2
                            existing_states.append((new_hall, row1, new_row2, new_total_cost_so_far, num_moves + 1))
                            if debug:
                                print("moved it")
                            continue
                        elif c == "C":
                            new_row2 = row2[:i] + ["."] + row2[i+1:]
                            new_total_cost_so_far = total_cost_so_far + 100 * abs([2, 4, 6, 8][i] - hall_pos) + 100 * 2
                            existing_states.append((new_hall, row1, new_row2, new_total_cost_so_far, num_moves + 1))
                            if debug:
                                print("moved it")
                            continue
                        elif c == "D":
                            new_row2 = row2[:i] + ["."] + row2[i+1:]
                            new_total_cost_so_far = total_cost_so_far + 1000 * abs([2, 4, 6, 8][i] - hall_pos) + 1000 * 2
                            existing_states.append((new_hall, row1, new_row2, new_total_cost_so_far, num_moves + 1))
                            if debug:
                                print("moved it")
                            continue
            
```

```python
test_state_string = """
#############
#...B.......#
###B#C#.#D###
  #A#D#C#A#
  #########
  """

example_state_string = """
#############
#...........#
###B#C#B#D###
  #A#D#C#A#
  #########
  """

real_state_string = """
#############
#...........#
###C#D#A#B###
  #B#A#D#C#
  #########
  """

#state_string = test_state_string
#starting_cost = 40
#starting_num_moves = 1

#state_string = example_state_string
#starting_cost = 0
#starting_num_moves = 0

state_string = real_state_string
starting_cost = 0
starting_num_moves = 0

hall = [c for c in state_string.split("\n")[2][1:12]]
row1 = [state_string.split("\n")[3][3], state_string.split("\n")[3][5], state_string.split("\n")[3][7], state_string.split("\n")[3][9]]
row2 = [state_string.split("\n")[4][3], state_string.split("\n")[4][5], state_string.split("\n")[4][7], state_string.split("\n")[4][9]]

state = (hall, row1, row2, starting_cost, starting_num_moves)

existing_states = [state]

print_states = False

print_strings = False
debug = False

num_cycles = 0
while True:
    num_cycles += 1
    
    new_states = []
    
    prev_states = existing_states[:]
    existing_states = []
    
    for s in prev_states:
        if print_strings:
            print_state(s)
        make_moves(s, debug=debug)

    if num_cycles % 1 == 0:
        print("\nstarting infrequent tasks")
        print("depth =", num_cycles)
        print("currently have", len(existing_states), "states")
        possible_costs = [x[3] for x in existing_states]
        if len(possible_total_costs) > 0:
            print("Min cost:", np.min(possible_total_costs))

            
        prev_states = existing_states[:]
        existing_states = []
        existing_state_strs = set()
                
        for s in prev_states:
            if not (str(s) in existing_state_strs):
                existing_states.append(s)
                existing_state_strs.add(str(s))
        print("removing duplicates, now", len(existing_states), "states")

        """if len(possible_total_costs) > 0:
            new_existing_states = []
            for s in existing_states:
                cost = s[4]
                if not (cost >= np.min(possible_total_costs)):
                    new_existing_states.append(s)
            print("removing high costs, now", len(new_existing_states), "states")
            existing_states = new_existing_states"""

        """new_existing_states = []
        for s in existing_states:
            num_moves = s[4] + 1
            num_times_added = int(1000. / num_moves)
            for _ in range(num_times_added):
                new_existing_states.append(s)
        print("adding duplicates now (making it choose less deep options more frequently)", len(new_existing_states))    
        existing_states = new_existing_states"""

        if print_states:
            print(f"printing {len(existing_states)} states...")
            for state in existing_states:
                print_state(state)
            print("done printing states")

        for state in existing_states:
            hall = state[0]
            row1 = state[1]
            row2 = state[2]
            total_cost_so_far = state[3]
            num_moves = state[4]

            if row1 == ["A", "B", "C", "D"] and row2 == ["A", "B", "C", "D"]:
                possible_total_costs.append(total_cost_so_far)
                #print("finished one!")

        print("done doing infrequent tasks")
```
