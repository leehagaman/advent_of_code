```python
with open("input_day_19.txt") as file:
    
    strs = [x.strip() for x in file.readlines()]
    
    beacon_coords = []
    for x in strs:
        if x == "":
            continue
        elif x[:2] == "--":
            beacon_coords.append([])
        else:
            beacon_coords[-1].append([int(_) for _ in x.split(",")])

#print(strs)
#print(beacon_coords)

```

```python
def merge_best(beacon_coords, scanner_coords):
    print("currently", len(beacon_coords), "beacons left")
    #print(scanner_coords)
    #print(beacon_coords)
    
    if len(beacon_coords) == 1:
        print("\ndone!")
        print("part a:", len(beacon_coords[0]))
        max_d = 0
        for x in scanner_coords:
            for y in scanner_coords:
                d = np.abs(x[0] - y[0]) + np.abs(x[1] - y[1]) + np.abs(x[2] - y[2])
                max_d = max(max_d, d)
                
        print("part b:", max_d)
        print(beacon_coords)
        print(len(scanner_coords), scanner_coords)
        return
    
    rel_dists = [] # list of relative distances between pairs in a single beacon
    rel_dists_coords = [] # pairs of 3D coordinates for each pair in a single beacon (same order as above)

    # get rel dists for each beacon
    for k in range(len(beacon_coords)):
        rel_dists.append([])
        rel_dists_coords.append([])

        for i in range(len(beacon_coords[k])):
            x = beacon_coords[k][i]
            for j in range(len(beacon_coords[k])):
                if i == j:
                    continue
                y = beacon_coords[k][j]

                rel_dists[-1].append((x[0] - y[0]) * (x[0] - y[0]) + (x[1] - y[1]) * (x[1] - y[1]) + (x[2] - y[2]) * (x[2] - y[2]))
                rel_dists_coords[-1].append([[x[0], x[1], x[2]], [y[0], y[1], y[2]]])

    max_num_dist_intersections = -1
    index_max_num_dist_intersections_i = 0
    index_max_num_dist_intersections_j = 0

    for i in range(1): # always use the first thingy, used to loop here
        a = rel_dists[i]
        for j in range(len(rel_dists)):
            if i == j:
                continue
                
            b = rel_dists[j]

            num_dist_intersections = len([_ for _ in a if _ in b])
            if num_dist_intersections > max_num_dist_intersections:
                max_num_dist_intersections = num_dist_intersections
                index_max_num_dist_intersections_i = i # should always be 0
                index_max_num_dist_intersections_j = j
                
                #print(i, j, "should be different")

    #print("best intersection is scanner", index_max_num_dist_intersections_i, "with scanner", index_max_num_dist_intersections_j, "with", max_num_dist_intersections, "distance overlaps\n")

    a_dists = rel_dists[index_max_num_dist_intersections_i]
    b_dists = rel_dists[index_max_num_dist_intersections_j]

    a_coords = rel_dists_coords[index_max_num_dist_intersections_i]
    b_coords = rel_dists_coords[index_max_num_dist_intersections_j]

    num_times = {}

    for i in range(len(a_dists)):
        a_dist = a_dists[i]
        a_coord = a_coords[i] # pair of coords
        for j in range(len(b_dists)):
            b_dist = b_dists[j]
            b_coord = b_coords[j]

            if a_dist == b_dist:
                if (i, j) in num_times:
                    num_times[(i, j)] += 1
                else:
                    num_times[(i, j)] = 1

    matching_coord_pair_and_num_times_same_dist = []
    lowest_num_times = 1000000000
    for k, v in num_times.items():
        matching_coord_pair_and_num_times_same_dist.append([[a_coords[k[0]], b_coords[k[1]]], v])

    matching_coord_pair_and_num_times_same_dist.sort()
    #print(matching_coord_pair_and_num_times_same_dist[:3])

    #print("most unique matching coord pair,", most_unique_matching_coord_pair, ", with ", lowest_num_times, "appearances")

    for p in range(len(matching_coord_pair_and_num_times_same_dist)):
        curr_coord_pair = matching_coord_pair_and_num_times_same_dist[p][0]


        #print("trying", curr_coord_pair)

        for swap in [0, 1]:
            a_coordinate = curr_coord_pair[0][0]
            b_coordinate = curr_coord_pair[1][swap]

            #print("original key a", a_coordinate)


            for perm_num in range(6):
                perm = [[0, 1, 2], [1, 2, 0], [2, 0, 1], [0, 2, 1], [2, 1, 0], [1, 0, 2]][perm_num]

                for x_sign in [-1, 1]:
                    for y_sign in [-1, 1]:
                        for z_sign in [-1, 1]:

                            flip = 1
                            if perm_num > 2:
                                flip = flip * -1
                            if x_sign == -1:
                                flip = flip * -1
                            if y_sign == -1:
                                flip = flip * -1
                            if z_sign == -1:
                                flip = flip * -1    
                            if flip == -1:
                                continue

                            #print("    original key b", b_coordinate)

                            transformed_key_b_coordinate = [x_sign * b_coordinate[perm[0]] + a_coordinate[0] - x_sign * b_coordinate[perm[0]],
                                                 y_sign * b_coordinate[perm[1]] + a_coordinate[1] - y_sign * b_coordinate[perm[1]],
                                                 z_sign * b_coordinate[perm[2]] + a_coordinate[2] - z_sign * b_coordinate[perm[2]]]


                            #print("    transformed key b", b_coordinate)

                            num_intersections = 0
                            
                            #print("new try")
                            
                            for a_coord in beacon_coords[index_max_num_dist_intersections_i]:
                                #print("        a:", a_coord)
                                
                                combined_a_b_coords = beacon_coords[index_max_num_dist_intersections_i][:]

                                for b_coord in beacon_coords[index_max_num_dist_intersections_j]:
                                    #print("    potentially adding b")
                                    transformed_b_coord = [x_sign * b_coord[perm[0]] + a_coordinate[0] - x_sign * b_coordinate[perm[0]],
                                                         y_sign * b_coord[perm[1]] + a_coordinate[1] - y_sign * b_coordinate[perm[1]],
                                                         z_sign * b_coord[perm[2]] + a_coordinate[2] - z_sign * b_coordinate[perm[2]]]
                                    #print("        original b:", b_coord)
                                    #print("        transformed b:", transformed_b_coord)

                                    if (a_coord == transformed_b_coord):
                                        num_intersections += 1
                                        #print("matching point!")
                                    elif transformed_b_coord not in combined_a_b_coords:
                                        combined_a_b_coords.append(transformed_b_coord)
                                        #print("        adding b")

                            if num_intersections >= 12:
                                #print(f"found {num_intersections} intersections!")
                                
                                #print("combined length", len(combined_a_b_coords))
                                #print(combined_a_b_coords)
                                
                                new_scanner_coords = scanner_coords + [[a_coordinate[0] - x_sign * b_coordinate[perm[0]],
                                                                       a_coordinate[1] - y_sign * b_coordinate[perm[1]],
                                                                       a_coordinate[2] - z_sign * b_coordinate[perm[2]]]]
                                
                                new_beacon_coords = [combined_a_b_coords]
                                for i in range(len(beacon_coords)):
                                    #print("potentially adding new beacon")
                                    if i != index_max_num_dist_intersections_i and i != index_max_num_dist_intersections_j:
                                        new_beacon_coords.append(beacon_coords[i])
                                        #print("adding untouched array")
                                        
                                return merge_best(new_beacon_coords, new_scanner_coords)
                            
    print("no intersection was good enough!!!")

scanner_coords = [[0,0,0]] # specifying coord for first one
merge_best(beacon_coords, scanner_coords)

print()
```
