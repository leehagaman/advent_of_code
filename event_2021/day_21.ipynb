```python
num_wins_player_1 = 0
num_wins_player_2 = 0

#state_dic = {(1, 4, 8, 0, 0, 0, 1) : 1}
state_dic = {(1, 8, 3, 0, 0, 0, 1) : 1}

while len(state_dic) > 0:
    
    #print(num_wins_player_1)
    #print(num_wins_player_2)
    #print(state_dic)
    
    
    new_state_dic = {}

    for tup, num in state_dic.items():
        
        for r1 in range(1, 4):
            for r2 in range(1, 4):
                for r3 in range(1, 4):
                    
                    turn = tup[0]
                    player_1_pos = tup[1]
                    player_2_pos = tup[2]
                    player_1_score = tup[3]
                    player_2_score = tup[4]
                    num_rolls = tup[5]
                    die_roll_num = tup[6]
                    
                    
                    num_spaces_to_move = r1 + r2 + r3
                    #print(num_spaces_to_move)
                    num_rolls += 3
                    
                    if turn == 1:
                        player_1_pos += num_spaces_to_move
                        while player_1_pos > 10:
                            player_1_pos -= 10
                        player_1_score += player_1_pos
                        
                        new_tup = (2, player_1_pos, player_2_pos, player_1_score, player_2_score, num_rolls, die_roll_num)
                        
                        if player_1_score >= 21:
                            num_wins_player_1 += num
                        elif new_tup in new_state_dic:
                            new_state_dic[new_tup] += num
                        else:
                            new_state_dic[new_tup] = num
                            
                    if turn == 2:
                        player_2_pos += num_spaces_to_move
                        while player_2_pos > 10:
                            player_2_pos -= 10
                        player_2_score += player_2_pos
                        
                        new_tup = (1, player_1_pos, player_2_pos, player_1_score, player_2_score, num_rolls, die_roll_num)
                        
                        if player_2_score >= 21:
                            num_wins_player_2 += num
                        elif new_tup in new_state_dic:
                            new_state_dic[new_tup] += num
                        else:
                            new_state_dic[new_tup] = num
                            
    state_dic = new_state_dic

print(num_wins_player_1)
print(num_wins_player_2)
```
